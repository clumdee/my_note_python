```python
import numpy as np
import scipy as sp
import pandas as pd
# import scipy.special as spc
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.ticker as mtick
```

```python
x = np.array(range(100))
y_1 = np.sin(x/10)
y_2 = np.cos(x/10)
```

```python
plt.rcParams['figure.figsize'] = 15, 6

# import seaborn as sns
# sns.set_style(None)
```

```python
fig, fig_ax = plt.subplots(1,2,figsize=(14,6),sharey=True)

for ax, data, title in zip(fig_ax, [(x,y_1),(x,y_2)], ['sin function','cos function']):
        
    ax.plot(data[0], data[1], label='default')
    ax.plot(data[0], 0.5+data[1], label='shifted')
    ax.set_title(title, fontsize=20)
    ax.set_xlabel('x', fontsize=18)
    ax.set_ylabel('function', fontsize=18)
    ax.set_xlim(0,100)
#     ax.set_xticks(ax.get_xticks()[0::2])
#     ax.set_xticklabels([])
    ax.set_ylim(-1.6,1.6)
#     ax.set_yticklabels([])

    ax.tick_params(direction='in',labelsize=16)
    ax.legend(loc=4, fontsize=20)
    ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.1f'))

fig_ax[0].scatter([20,40],[-0.5,1])
fig_ax[0].annotate('text1', xy=(20, -0.5), xytext=(20+2, -0.5), fontsize=14)   

plt.show()
```
